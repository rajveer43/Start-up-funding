#### Extracting some of the useful data

```r
dataset <- read.csv('/content/startup data.csv')

dataset <- select(dataset, founded_at, funding_rounds, funding_total_usd, is_software,
                       is_web,is_mobile,is_enterprise,is_advertising,is_gamesvideo
                       ,is_ecommerce,is_biotech,is_consulting,is_othercategory,
                       has_VC, has_roundA, has_roundB, has_roundC, avg_participants, status)

print(colnames(dataset))
print(ncol(dataset))
print(nrow(dataset))
```

```r
glimpse(dataset)
```

#### Formatting the data and converting it to proper datatype.

```r

dataset$status[dataset$status=='acquired'] <- 1
dataset$status[dataset$status=='closed'] <- 0


dataset$status <- as.integer(dataset$status)
dataset$founded_at <- as.Date(dataset$founded_at, "%m/%d/%y")
```

```r
glimpse(dataset)
```

### Splitting the dataset in train and test set

```r
set.seed(100)
split = sample.split(dataset$status, SplitRatio = 0.8)
trainData = subset(dataset, split==TRUE)
testData = subset(dataset, split==FALSE)
```

### Building a Random Forest to classify the the success to a given StartUp.

```r
set.seed(10)
trainData$status = as.factor(trainData$status)
testData$status = as.factor(testData$status)

RF_model = randomForest(status ~ ., data=trainData, mtry=5)
predictRF = predict(RF_model, newdata=testData)
print("!!!! Confusion Matrix !!!!")
table(testData$status, predictRF)
```
